Observe that each duplicate ID appears to have a row with what seems like valid data, and one row that's entirely zeros. A row of all zeros is invalid data, as it makes no sense for someone to have an age of 0, credit line of 0 and so on. 

One approach is to find rows that have all zeros, except for the first column, which has the IDs. We can find entries of the DataFrame that are zero with the conditional, "is equal to zero", this time on the entire dataframe, to yield a Boolean matrix, the same size as the DataFrame.

The goal is to create a Boolean Series, that identifies every row where all the elements, starting from the second column are 0. Use the .iloc[] method select all columns but the first one. We use .iloc[] (aka integer locator) so that we can select all but the first column using 1:  (ie an integer) as the second argument to iloc. We then apply the all() method along the column axis (axis 1). The all() method checks if all entries are True over an axis. By so doing we will select all rows which have only 0 entries from the second column onwards. The all() method yields a boolean mask which we sum to find that 315 rows have all zero entries from the second column onwards. This includes the 313 duplicates and two other non duplicates. We started out looking for duplicates and ended up removing invalid rows.

Finally with our boolean Series mask, we invert using ~, and use .loc[] (since we are using row labels, and : for columns) to select all the non duplicate, valid rows. We also use the .copy() method to create a copy of this output, instead of a view on the original DataFrame. We also store the resulting dataframe to csv (excluding the integer index values) as its a good idea to save intermediate data as we make significant changes.

```python
df_zero_mask = df == 0  # boolean matrix, telling us if the value in any cell is 0 or not (True / False)
```

```python
feature_zero_mask = df_zero_mask.iloc[:, 1:].all(axis=1) # if from second column, all are 0, new boolean mask
```

```python
sum(feature_zero_mask)  # total number of rows having all zero columns (2 non duplicates had all zeros)
```

```python
df_clean_1 = df.loc[~feature_zero_mask,:].copy()  # invert with ~ to choose rows that are not all zeroes
```

After importing our intermediate cleaned data, we notice that PAY_1 payment status column has the generic object data type, all the others have int64 data type so we investigate further.

The data dictionary states that the valid valued in PAY_1 column (payment status of the most recent month's bill) are -1,1,2,3, and so on.  According to the data dictionary, the measurement scale of the repayament status is: -1=pay duly; 1=payment delay for one month; 2=payment delay for two months;...;8=payment delay for eight months;9=payement delay for nine months and above.The preceding output reveals the prescence of two undocumented values: 0 and -2; 

There is a 'Not available' string present in this column (symbolizing missing data), which is the reason why this column was imported as a object data type, instead of int64. First we create a boolean mask to find all rows that don't have a missing data in PAY_1 feature, sum the boolean mask to find the rows that have available data and then remove the rows with 'Not available' using the .loc[] slice and .copy() to make sure we have a new dataframe not just a view. We still have PAY_1 values 0 and -2 which are not defined in the data dictionary. To be consistent, we also cast the data type of PAY_1 as int64 using the .astype method.

Now after the second data cleaning operation, we find out that undocumented values of -2 and 0 in PAY_1 have valid explanations.
- -2 means the account started that month with a zero balance and never used any credit.
- -1 means the account had a balance that was paid in full.
- 0 means that at least the minimum payment was made, but the entire balance wasn't paid (that is, a positive balance was carried to the next month).

We save the results of the second data cleaning operation as its a good idea to save intermediate data after each data cleaning operation.

#### Ordinal Categorical Features.

Some classification algorithms such as logistic regression assume a linear effect of the features. For binary classification, you can look at the different levels of a categorical feature in terms of the average values of the response variable within each level. For regression problems, we would model the average response variable within each level.

Below we use the groupby()/ agg procedure, followed by plot.bar to plot a bar plot of the mean default payment next month per education level.

From looking at the bar plot, a straight-line fit would probably not be the best description of the data here. In case a feature has a non-linear effect like this, it may be better to use a non-linear algorithm such as a decision tree or random forest. If a simpler, and more interpretable linear model such as logistic regression is desired, we could avoid an ordinal encoding and use a different way of encoding categorical variables called one-hot encoding (OHE).

```python
df_clean_2 = pd.read_csv('intermediate_data/df_clean_2_01.csv')
df_clean_2.groupby('EDUCATION').agg({'default payment next month': 'mean'})\
                                .plot.bar(legend=False,
                                          figsize=(12,8),
                                          grid='both'
                        
                                          ) # plot mean default payment next month per level
plt.ylabel('Default rate')
plt.xlabel('Education level : ordinal encoding')
plt.tight_layout()
```

#### Implementing One-Hot Encoding for a Categorical Feature

```python
# Create an empty column for the categorical labels 
df_clean_2['EDUCATION_CAT'] = 'none'
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
# Create a mapping dictionary
cat_mapping = {1: "graduate school",
                2: "university",
                3: "high school",
                4: "others"}
```

```python
# apply mapping
df_clean_2["EDUCATION_CAT"] = df_clean_2['EDUCATION'].map(cat_mapping)
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
edu_ohe = pd.get_dummies(df_clean_2['EDUCATION_CAT'], dtype=int)
edu_ohe.head(10)
```

```python
# Concatenate the one-hot encoded DataFrame to the original DataFrame
df_with_ohe = pd.concat([df_clean_2, edu_ohe], axis=1) # to concatenate along axis 1, must have same row ids
df_with_ohe[['EDUCATION_CAT', 'graduate school',\
              'high school', 'university', 'others']].head(10)
```

```python
df_with_ohe.to_csv('intermediate_data/Chapter_1_cleaned_data.csv', index=False)
```

Something isn't write in the histograms for payment statuses from September to April ie PAY_1 to PAY_6. Specifically, there are barely any accounts that are one month late in these histograms but there a bunch of accounts  are two months late. The number of accounts with a 2-month delay (encoded as 2) in a given month should be less than or equal to the number of accounts with a 1-month delay in the previous month (encoded as 1). For example if an account is two-month late in August, then it follows logically that, the same account was 1-month late in July. 

This means the payment statuses for August to April are not representative of future data and such cannot be used to build our predictive model. Only PAY_1 is representative of future data, since the value of 1 is underreported in all months except for September. This shows the importance of a thorough examination of data quality and the utility of visualization for such examination.
